```python
from statsmodels.tsa.seasonal import seasonal_decompose
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from pmdarima import auto_arima
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.dates as mdates
import scipy.stats
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pylab
sns.set(style='white')
```

```python
df = pd.read_csv('../datasets/inidices/NIFTY 50.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(['Date'], inplace=True)
```

```python
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df=df.loc['2000-01-03':]
```

```python
fig = go.Figure([go.Scatter(x=df.index, y=df['Turnover'])])
fig.update_layout(
    autosize=False,
    width=1000,
    height=500,
    title='Turnover over time',
    template="simple_white",
)
fig.update_xaxes(title="Date")
fig.update_yaxes(title="Turnover")
fig.show()
```

```python
fig = go.Figure(
    [go.Scatter(x=df.loc['2008', 'Volume'].index, y=df.loc['2009', 'Volume'])])
fig.update_layout(
    autosize=False,
    width=1000,
    height=500,
    title='Volume in 2008',
    template="simple_white",
)
fig.update_xaxes(title="Date")
fig.update_yaxes(title="Volume")

fig.show()
```

```python
fig = go.Figure(go.Candlestick(x=df.index, open=df.Open, high=df.High, low=df.Low, close=df.Close))
fig.show()
```
